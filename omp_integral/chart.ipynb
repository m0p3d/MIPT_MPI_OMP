```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('times.csv', sep='\t')
n = len(data)
data.head()
```

```python
plt.figure(figsize=(20, 5))
plt.plot(np.arange(1, n + 1), data['critical'][0] / data['critical'])
plt.title('Ускорение программы, использующей критическую секцию')
plt.xticks(np.arange(1, n))
plt.yticks(np.arange(1, max(data['critical'][0] / data['critical'])+1, 0.5))
plt.grid(True)
plt.xlabel("Количество нитей")
plt.ylabel("Ускорение")
plt.show()
```

```python
plt.figure(figsize=(20, 5))
local_data = data['reduction']
plt.plot(np.arange(1, n + 1), local_data[0] / local_data)
plt.title('Ускорение программы, использующей reduction')
plt.xticks(np.arange(1, n))
plt.yticks(np.arange(1, max(local_data[0] / local_data) + 1, 0.5))
plt.grid(True)
plt.xlabel("Количество нитей")
plt.ylabel("Ускорение")
plt.show()
```
