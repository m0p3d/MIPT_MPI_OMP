```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('times.csv', sep='\t')
n = len(data)
data.head()
```

```python
plt.figure(figsize=(20, 5))
plt.plot(np.arange(1, n + 1), data['time'][0] / data['time'])
plt.title('Ускорение программы')
plt.xticks(np.arange(1, n))
plt.yticks(np.arange(1, max(data['time'][0] / data['time'])+1, 0.5))
plt.grid(True)
plt.xlabel("Количество нитей")
plt.ylabel("Ускорение")
plt.show()
```

```python
plt.figure(figsize=(20, 5))
plt.plot(np.arange(1, n + 1), data['time'][0] / data['time'] / np.arange(1, n + 1))
plt.title('Эффективность программы')
plt.xticks(np.arange(1, n))
plt.yticks(np.arange(1, max(data['time'][0] / data['time'])+1, 0.5))
plt.grid(True)
plt.xlabel("Количество нитей")
plt.ylabel("Ускорение")
plt.show()
```
